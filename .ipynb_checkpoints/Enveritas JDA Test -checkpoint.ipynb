```python
import numpy as np
import  pandas as pd
```

```python
data=pd.read_excel("Enveritas JDA Test Workbook and Instructions_EA.xlsx",  sheet_name="Task 3", header=2)
data.head()
```

### Q1. Average Yield in the three regions

```python
average_yield_per_region=data.groupby("Coffee Region")["Yield (kg/green) per hectare"].mean().reset_index()
average_yield_per_region
```

### Q2.Farm management practices mostly used among farmers

##### i. Use of Disease Resistant Coffee Varieties

```python
diseases_resistant_variety = data.groupby('Coffee Region')['Use Disease Resistant Coffee Varieties'].value_counts()
diseases_resistant_variety
```

##### ii. Use Organic Fertilizer

```python
organic_fertilizer = data.groupby('Coffee Region')['Use Organic Fertilizer'].value_counts()
organic_fertilizer = pd.crosstab(data['Coffee Region'], data['Use Organic Fertilizer'])
organic_fertilizer
```

##### iii. Intercrop food tree with coffee

```python
intercropping = data.groupby('Coffee Region')['Intercrop food tree with coffee'].value_counts()
intercropping = pd.crosstab(data['Coffee Region'], data['Intercrop food tree with coffee'])
intercropping
```

#### iv. Number of times weeded in a year

```python
intercropping = data.groupby('Coffee Region')['Number of times farmer weeded in last year'].value_counts()
intercropping = pd.crosstab(data['Coffee Region'], data['Number of times farmer weeded in last year'])
intercropping
```

### Q3. Prunning techniques that are mostly used

```python
prunning_techniques=data['Pruning techniques used'].value_counts().reset_index()
prunning_techniques
```
